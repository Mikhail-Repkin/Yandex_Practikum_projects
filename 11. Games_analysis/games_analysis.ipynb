–í —Ö–æ–¥–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –±—ã–ª –ø–æ–ø—Ä–∞–≤–ª–µ–Ω —Å—Ç–∏–ª—å –∑–∞–≥–æ–ª–æ–≤–∫–æ–≤, —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö `name` –∏ `genre` –±—ã–ª–∏ —É–¥–∞–ª–µ–Ω—ã, —á–∞—Å—Ç—å –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ `year_of_release` –±—ã–ª–∞ –∑–∞–ø–æ–ª–Ω–µ–Ω–∞ –≥–æ–¥–æ–º, –∏–∑–≤–ª–µ—á–µ–Ω–Ω—ã–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `name`, –¥—Ä—É–≥–∞—è —á–∞—Å—Ç—å —É–¥–∞–ª–µ–Ω–∞.     
–ü—Ä–æ–ø—É—Å–∫–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏—è tbd –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö `critic_score` –∏ `user_score` –±—ã–ª–∏ –∑–∞–ø–æ–ª–Ω–µ–Ω—ã –∑–Ω–∞—á–µ–Ω–∏–µ–º-–∑–∞–≥–ª—É—à–∫–æ–π. –ü—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ `rating` –∑–∞–ø–æ–ª–Ω–µ–Ω—ã –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–π –∑–∞–≥–ª—É—à–∫–æ–π 'unknown'.    
–ö—Ä–æ–º–µ —Ç–æ–≥–æ, –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –±—ã–ª –¥–æ–±–∞–≤–ª–µ–Ω –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏, –∞ –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –∫ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–º—É —Ç–∏–ø—É.

## –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –ê–Ω–∞–ª–∏–∑ –≤—ã—Ö–æ–¥–∞ –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è bar chart
def bar_chart(dataframe, grouping_column, total_column):
             
    # –Ω–∞–∑–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
    chart_name={'year_of_release':'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã—à–µ–¥—à–∏—Ö –∏–≥—Ä –ø–æ –≥–æ–¥–∞–º (—à—Ç.)'
               }

    # –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞
    df_filtered = dataframe.groupby([grouping_column])[total_column].count()          
                
    # barchart –¥–ª—è –∏—Å–∫–æ–º–æ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è
    ax = df_filtered.plot(kind='bar',
                          y=total_column,
                          title=chart_name[grouping_column],
                          figsize=(15,5),
                          legend=False,
                          color = ['#0AA1DD'],
                          xlabel = ""
                         )
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–ø–∏—Å–µ–π –±–∞—Ä–æ–≤
    ax.bar_label(ax.containers[0], label_type='edge', color='black', fontsize=12, padding=10, rotation=90)
    
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Å–µ–π
    ax.spines['top'].set_visible(False) #—Å–±—Ä–æ—Å –≥—Ä–∞–Ω–∏—Ü –≥—Ä–∞—Ñ–∏–∫–∞
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    plt.yticks([]) #—Å–±—Ä–æ—Å y-–æ—Å–∏
```

```python
# –±–∞—Ä—á–∞—Ä—Ç 
plt.ylim(0,1800)
bar_chart(df_games, 'year_of_release', 'name')
```

–í–≤–∏–¥—É –º–∞–ª–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—ã—à–µ–¥—à–∏—Ö –∏–≥—Ä –≤ 1980-1993 –≥. –º–æ–∂–Ω–æ –∑–∞–∫–ª—é—á–∏—Ç—å —á—Ç–æ –∏–Ω—Ç–µ—Ä–µ—Å –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç –¥–∞–Ω–Ω—ã–µ –æ–± –∏–≥—Ä–∞—Ö –ø–æ—Å–ª–µ 1994 –≥. 

### –ê–Ω–∞–ª–∏–∑ –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º

–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º, –∫–∞–∫ –º–µ–Ω—è–ª–∏—Å—å –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º. –°–Ω–∞—á–∞–ª –æ–ø—Ä–µ–¥–µ–ª–∏–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏.

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è bar chart
def bar_chart_h(dataframe, grouping_columns, total_column):
             
    # –Ω–∞–∑–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
    chart_name={'platform':'–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º (–≤ –º–ª–Ω. –∫–æ–ø–∏–π)'
               }

    # –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞
    df_filtered = dataframe.groupby(grouping_columns)[total_column].sum().sort_values()          
    
    # barchart –¥–ª—è –∏—Å–∫–æ–º–æ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è
    ax = df_filtered.plot(kind='barh',
                          y=total_column,
                          title=chart_name[grouping_columns],
                          figsize=(15,8),
                          width=0.8,
                          legend=False,
                          color = ['#0AA1DD'],
                          xlabel = ""
                         )
   
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–ø–∏—Å–µ–π –±–∞—Ä–æ–≤
    ax.bar_label(ax.containers[0], label_type='edge', color='black', fontsize=10, padding=10)
    
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Å–µ–π
    ax.spines['top'].set_visible(False) #—Å–±—Ä–æ—Å –≥—Ä–∞–Ω–∏—Ü –≥—Ä–∞—Ñ–∏–∫–∞
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    plt.xticks([]) #—Å–±—Ä–æ—Å —Ö-–æ—Å–∏
    
    # —Ç–∞–±–ª–∏—Ü–∞-—Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º
    top_sales = df_filtered.sort_values(ascending=False).reset_index()[grouping_columns].tolist()
    
    return top_sales
```

```python
# –±–∞—Ä—á–∞—Ä—Ç 
plt.xlim(0,1500)
top_sales = bar_chart_h(df_games, 'platform', 'total_sales')
```

–ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–Ω–∞–º–∏–∫—É –ø—Ä–æ–¥–∞–∂ –ø–æ –≥–æ–¥–∞–º –ø–æ —Ç–æ–ø-6 –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º:

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è line chart
def line_chart(dataframe, top_platforms):

    for platform in top_platforms:

        # –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞
        df_filtered = dataframe.groupby(['platform','year_of_release']).sum().reset_index().query('platform == @platform')
    
        # barchart –¥–ª—è –∏—Å–∫–æ–º–æ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è
        chart = df_filtered.plot(kind='line',
                         x='year_of_release',
                         y='total_sales',
                         title="–î–∏–Ω–∞–º–∏–∫–∞ –ø—Ä–æ–¥–∞–∂ –ø–æ –≥–æ–¥–∞–º –ø–æ –ª—É—á—à–∏–º –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º (–≤ –º–ª–Ω. –∫–æ–ø–∏–π)",
                         style='o-',
                         figsize=(15,7),
                         label=platform,
                         ylabel = "—Å—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ (–º–ª–Ω. —à—Ç.)",
                         xlabel = "–≥–æ–¥ –≤—ã–ø—É—Å–∫–∞",
                         ax = plt.gca() #–ø–æ–ª—É—á–µ–Ω–∏–µ —Ç–µ–∫—É—â–∏—Ö –æ—Å–µ–π
                        )
        chart.xaxis.set_major_locator(ticker.MultipleLocator(1)) #—É—Å—Ç–∞–Ω–∞–≤–∫–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –æ—Å–Ω–æ–≤–Ω—ã—Ö –¥–µ–ª–µ–Ω–∏–π
        chart.tick_params(labelrotation = 90) #–ø–æ–≤–æ—Ä–æ—Ç –ø–æ–¥–ø–∏—Å–µ–π
```

### Boxplot –ø–æ –≥–ª–æ–±–∞–ª—å–Ω—ã–º –ø—Ä–æ–¥–∞–∂–∞–º –∏–≥—Ä –≤ —Ä–∞–∑–±–∏–≤–∫–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ ¬´—è—â–∏–∫ —Å —É—Å–∞–º–∏¬ª –ø–æ –≥–ª–æ–±–∞–ª—å–Ω—ã–º –ø—Ä–æ–¥–∞–∂–∞–º –∏–≥—Ä –¥–ª—è –¥–∞–Ω–Ω—ã—Ö —Å 2004 –≥. –≤ —Ä–∞–∑–±–∏–≤–∫–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∏ –æ–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç:

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è box plot
def boxplot_platforms(dataframe, top_platforms, total_column, ylim=None):
    
    """
    dataframe - –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
    top_platforms - —Å–ø–∏—Å–æ–∫ –ø–ª–∞—Ç—Ñ–æ—Ä–º
    total_column - —Ü–µ–ª–µ–≤–æ–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å (y-–æ—Å—å)
    """
    
    # –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞
    df_filtered = dataframe.query('platform == @top_platforms')
    
    # boxplot 
    ax = df_filtered.boxplot(column=total_column, by='platform', fontsize=15, figsize=(18,8))
    
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–ø–∏—Å–µ–π –≤ boxplot 
    ax.set_xlabel('') #—É–±—Ä–∞—Ç—å –Ω–∞–∑–≤–∞–Ω–∏–µ x-–æ—Å–∏
    ax.set_title('Boxplot –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –≤ —Ä–∞–∑–±–∏–≤–∫–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º', fontsize=18) #–Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
    plt.suptitle('') #—É–±—Ä–∞—Ç—å –ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–æ–∫
    ax.set_ylim(ylim) #–Ω–∞—Å—Ç—Ä–æ–π–∫–∞ y-–æ—Å–∏
```

```python
# boxplot
boxplot_platforms(df_games, top_sales_from_2004, 'total_sales')
```

```python
# boxplot (—É–∫—Ä—É–ø–Ω—ë–Ω–Ω–æ)
boxplot_platforms(df_games, top_sales_from_2004, 'total_sales', [0,2])
```

Wii –∏–º–µ–µ—Ç –≤ —Å–≤–æ–µ–º –∞—Ä—Å–µ–Ω–∞–ª–µ –≤—ã–¥–∞—é—â—É—é—Å—è –∏–≥—Ä—É ‚Äî Wii Sports, –∫–æ—Ç–æ—Ä–∞—è –æ—Ç–º–µ—á–µ–Ω–∞ 82.5 –º–ª–Ω. –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π.

### –ê–Ω–∞–ª–∏–∑ –≤–ª–∏—è–Ω–∏—è –æ—Ç–∑—ã–≤–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏

–í –∫–∞—á–µ—Å—Ç–≤–µ –ø–æ–ø—É–ª—è—Ä–Ω–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –≤—ã–±–µ—Ä–µ–º X360 ‚Äî –ª–∏–¥–µ—Ä–∞ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∏–≥—Ä –≤ 2004-2016 –≥–≥.

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
def scatter_plot(dataframe, y, x):
    
    """
    dataframe - –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
    y - —Å—Ç–æ–ª–±–µ—Ü —Å –∑–∞–≤–∏—Å–∏–º—ã–º–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ (y-–æ—Å—å)
    x - —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (x-–æ—Å—å)
    """
    
    # –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ü–∏—Ä—Å–æ–Ω–∞
    display(dataframe[[y, x]].corr())

    # –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏
    dataframe.plot(kind='scatter',
                     y=y,
                     x=x,
                     title=f'–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å {y} –æ—Ç {x}', 
                     figsize=(15,5),
                     fontsize=12,
                     legend=True)
```

–ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –æ–±—â–µ–π —Å—É–º–º—ã –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –æ—Ç –æ—Ç–∑—ã–≤–æ–≤ –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ –ø–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ—è–Ω–∏—è:

### –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–≥—Ä –ø–æ –∂–∞–Ω—Ä–∞–º

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—ã—à–µ–¥—à–∏—Ö –∏–≥—Ä –∏ —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –ø–æ –∂–∞–Ω—Ä–∞–º, –≤—ã–¥–µ–ª–∏–º –Ω–∞–∏–±–æ–ª–µ–µ –∏ –Ω–∞–∏–º–µ–Ω–µ–µ –∞–∫—Ç–∏–≤–Ω—ã–µ –∂–∞–Ω—Ä—ã.

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è bar chart
def bar_chart_combined(dataframe, grouping_column):
    
    ax = plt.gca()
    
    # –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞
    df_filtered = dataframe.groupby([grouping_column]).agg(
        {'name': 'count', 'total_sales': 'sum'}).round(0).sort_values(
        by='name', ascending=False)
       
    # barcharts
    barchart_1 = df_filtered['name'].plot(kind='bar',
                     figsize=(15,5),
                     color = '#0AA1DD',
                     width=0.3,
                     ax=ax,
                     position=1
                    )
    
    barchart_2 = df_filtered['total_sales'].plot(kind='bar',
                     figsize=(15,5),
                     color = '#F76E11',
                     width=0.3,
                     ax=ax,
                     position=0
                    )
    
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–ø–∏—Å–µ–π –±–∞—Ä–æ–≤
    ax.bar_label(ax.containers[0], label_type='edge', color='black', fontsize=10, padding=10, rotation=0)
    ax.bar_label(ax.containers[1], label_type='edge', color='black', fontsize=10, padding=10, rotation=0)
    
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Å–µ–π
    ax.spines['top'].set_visible(False) #—Å–±—Ä–æ—Å –≥—Ä–∞–Ω–∏—Ü –≥—Ä–∞—Ñ–∏–∫–∞
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    ax.legend(['–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã—à–µ–¥—à–∏—Ö –∏–≥—Ä (—à—Ç.)', '–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä (–º–ª–Ω. –∫–æ–ø–∏–π)']) #–ª–µ–≥–µ–Ω–¥–∞
    ax.set_title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã—à–µ–¥—à–∏—Ö –∏–≥—Ä (—à—Ç.) –∏ —Å—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –ø–æ –∂–∞–Ω—Ä–∞–º (–º–ª–Ω. –∫–æ–ø–∏–π)', fontsize=12) #–Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
    plt.yticks([]) #—Å–±—Ä–æ—Å y-–æ—Å–∏
```

```python
# –±–∞—Ä—á–∞—Ä—Ç 
plt.ylim(0,3200)
bar_chart_combined(df_games, 'genre')
```

–¢–∞–∫–∂–µ –¥–ª—è –ø–æ–∏—Å–∫–∞ –ø—Ä–∏–±—ã–ª—å–Ω–æ–≥–æ –∂–∞–Ω—Ä–∞ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø—Ä–æ–¥–∞–∂ –ø–æ –∫–∞–∂–¥–æ–º—É –∂–∞–Ω—Ä—É:

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è box plot
def boxplot_genres(dataframe, total_column, ylim=None):
    
    """
    dataframe - –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
    total_column - —Ü–µ–ª–µ–≤–æ–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å (y-–æ—Å—å)
    """
    
    # –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞
#    df_filtered = dataframe.query('platform == @top_platforms')
    
    # boxplot 
    ax = dataframe.boxplot(column=total_column, by='genre', fontsize=15, figsize=(18,8))
    
    # –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–ø–∏—Å–µ–π –≤ boxplot 
    ax.set_xlabel('') #—É–±—Ä–∞—Ç—å –Ω–∞–∑–≤–∞–Ω–∏–µ x-–æ—Å–∏
    ax.set_title('Boxplot –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –≤ —Ä–∞–∑–±–∏–≤–∫–µ –ø–æ –∂–∞–Ω—Ä–∞–º', fontsize=18) #–Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
    plt.suptitle('') #—É–±—Ä–∞—Ç—å –ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–æ–∫
    ax.set_ylim(ylim) #–Ω–∞—Å—Ç—Ä–æ–π–∫–∞ y-–æ—Å–∏
```

```python
# boxplot (—É–∫—Ä—É–ø–Ω—ë–Ω–Ω–æ)
boxplot_genres(df_games, 'total_sales', [0,2])
```

## –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ

–û–ø—Ä–µ–¥–µ–ª–∏–º –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ä–µ–≥–∏–æ–Ω–æ–≤ (NA, EU, JP) c–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∏ –∂–∞–Ω—Ä—ã (—Ç–æ–ø-5), –æ–ø–∏—à–µ–º —Ä–∞–∑–ª–∏—á–∏—è –≤ –¥–æ–ª—è—Ö –ø—Ä–æ–¥–∞–∂, –∏—Å—Å–ª–µ–¥—É–µ–º –≤–ª–∏—è–Ω–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∞ ESRB –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –≤ —Ä–µ–≥–∏–æ–Ω–∞—Ö.

```python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è pie chart
def pie_chart(dataframe, grouping_columns, region, ax):
    
    # –Ω–∞–∑–≤–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
    chart_name={'na_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –°–µ–≤. –ê–º–µ—Ä–∏–∫–µ',
                'eu_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –ï–≤—Ä–æ–ø–µ',
                'jp_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –Ø–ø–æ–Ω–∏–∏'}
    
    chart_name_add={'platform': ' –ø–æ —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º',
                    'genre':' –ø–æ —Ç–æ–ø-5 –∂–∞–Ω—Ä–∞–º',
                    'rating':' –ø–æ —Ä–µ–π—Ç–∏–Ω–≥–∞–º'}
        
    # –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞
    df_filtered = dataframe.groupby([grouping_columns]).agg({region:'sum'}).sort_values(by=region, ascending=False).head()
    
    # pie_chart
    chart = df_filtered.plot(kind='pie',
                     ax=axes[ax],
                     y=region,
                     autopct=lambda p: '{:.2f}%\n({:.0f})'.format(p,(p/100)*df_filtered[region].sum()),
                     title=chart_name[region] + chart_name_add[grouping_columns],
                     fontsize=12,
                     legend=False,
                     colors = sns.color_palette("Blues_r"), #_r - reversed color
                     ylabel = "")
```

```python
# –∫–æ—Ä—Ç–µ–∂—É –∏–∑ –¥–≤—É—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ figure –∏ axes –ø–µ—Ä–µ–¥–∞–µ–º –∫–æ–ª-—Ç–≤–æ —Å—Ç—Ä–æ–∫/—Å—Ç–æ–ª–±—Ü–æ–≤ —Å–µ—Ç–∫–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∞
fig, axes = plt.subplots(1, 3, figsize=(18,5))

# pie_charts
pie_chart(df_games, 'platform', 'na_sales', 0)
pie_chart(df_games, 'platform', 'eu_sales', 1)
pie_chart(df_games, 'platform', 'jp_sales', 2)
```

–í –Ø–ø–æ–Ω–∏–∏ –≤–µ—Å—å —Ç–æ–ø-5 –∑–∞–Ω—è—Ç –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º–∏ –º–µ—Å—Ç–Ω–æ–≥–æ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ ‚Äî –ª–∏–¥–∏—Ä—É—é—â–∏–º–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º–∏ —è–≤–ª—è–µ—Ç—Å—è —Å–µ–º–µ–π—Å—Ç–≤–æ Nintendo ‚Äî –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã DS –∏ 3DS  (34.9% –∏ 20.1% –ø—Ä–æ–¥–∞–∂), —Å–ª–µ–¥–æ–º –∏–¥—É—Ç –ø—Ä–∏—Å—Ç–∞–≤–∫–∏ Playstation.

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑

–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑ –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π –ø–æ –≤–∑—è—Ç—ã–º –≤—ã–±–æ—Ä–∫–∞–º –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å t-—Ç–µ—Å—Ç.

### –ì–∏–ø–æ—Ç–µ–∑–∞ –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC

**–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0)** ‚Äî —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã.    

**–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1)** ‚Äî —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC.

–ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π –ø–æ –≤–∑—è—Ç—ã–º –∏–∑ –Ω–∏—Ö –≤—ã–±–æ—Ä–∫–∞–º:

```python
# –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
# –µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –º–µ–Ω—å—à–µ –Ω–µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É
alpha = .05
```

```python
# —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–≤
df_games_XOne = df_games.query('(platform == "XOne") & (user_score > 0)')
df_games_PC = df_games.query('(platform == "PC") & (user_score > 0)')
```

```python
# –ø—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã
results_1st_hyp = st.ttest_ind(df_games_XOne['user_score'], df_games_PC['user_score'])

print('p-value:', results_1st_hyp.pvalue)

if results_1st_hyp.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```

**–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1)** ‚Äî —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports.

–ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π –ø–æ –≤–∑—è—Ç—ã–º –∏–∑ –Ω–∏—Ö –≤—ã–±–æ—Ä–∫–∞–º:
