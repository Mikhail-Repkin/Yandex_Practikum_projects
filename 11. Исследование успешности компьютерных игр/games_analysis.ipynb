```python
# удаление строк с пропущенными значениями
df_games = df_games.dropna(subset=['name']).reset_index(drop=True)
```

```python
# проверка количества пропущенных значений
print('Количество пропущенных значений:','{:.0f}'.format(df_games['name'].isna().sum()))
```

```python
df_games[df_games['year_of_release'].isna()].head()
```

```python
# подсчет пропусков в year_of_release
print('Количество пропусков:', len(df_games[df_games['year_of_release'].isna()]))
```

```python
# подсчет кол-ва строк одновременно содержащих год в колонке name и пропуск в year_of_release
len(df_games[(df_games['name'].str.contains(r'\d{4}')) & (df_games['year_of_release'].isna())])
```

В колонке year_of_release 269 пропусков. Однако некоторые значения в name содержат год выхода. Извлечем его из названий игр и заполним пропуски в year_of_release:

```python
# булевый список где колонка name содержит год, а year_of_release содержит пропуск 
name_year_bool = (df_games['name'].str.contains(r'\d{4}')) & (df_games['year_of_release'].isna())
```

```python
# используем булевый список для создания среза, где будем заполнять пропуски в year_of_release извлеченным из name годом
df_games.loc[name_year_bool, 'year_of_release'] = df_games[name_year_bool]['name'].str.extract(r'(\d{4})')[0]
```

```python
# подсчет пропусков
print('Количество пропусков:', len(df_games[df_games['year_of_release'].isna()]))
```

Количество пропусков стало на 17 штук меньше. Удалим оставшиеся пропущенные значения:

```python
# удаление строк с пропущенными значениями
df_games = df_games.dropna(subset=['year_of_release']).reset_index(drop=True)
```

```python
# подсчет количества пропущенных значений
print('Количество пропущенных значений:','{:.0f}'.format(df_games['year_of_release'].isna().sum()))
```

**Обработка пропусков в столбцах `critic_score` и `user_score`**

```python
# просмотр строк с пропущенными значениями
df_games[df_games['critic_score'].isna()].head()
```

```python
# подсчет количества пропущенных значений
print('Количество пропущенных значений в critic_score:','{:.0f}'.format(df_games['critic_score'].isna().sum()))
print('Количество пропущенных значений в user_score:','{:.0f}'.format(df_games['user_score'].isna().sum()))
```

💡 Доля пропусков в столбцах с оценками существенна — от 40% до 51%. Возможно, для части старых игр оценки не выставлялись в то время, либо информация о них не доступна. Кроме того для малопопулярных игр оценка может просто отсутствовать. В любом случае, количество пропусков слишком велико, что бы удалять данные строки — заменим их на значение-заглушку: -1.

💡 Кроме пропусков Nan, в столбце `user_score` содержатся значения tbd, что означает "To Be Determined" и проставляется для тех игр, по которым пока что недостаточно отзывов. Заменим tbd на Nan.

```python
# замена категориального значения на Nan
df_games['user_score'] = df_games['user_score'].replace('tbd', np.NaN)
```

```python
# замена пропусков на значение заглушку (-1)
df_games['user_score'] = df_games['user_score'].fillna(-1).astype('float32')
df_games['critic_score'] = df_games['critic_score'].fillna(-1).astype('float32')
```

```python
# подсчет количества пропущенных значений
print('Количество пропущенных значений в critic_score:','{:.0f}'.format(df_games['critic_score'].isna().sum()))
print('Количество пропущенных значений в user_score:','{:.0f}'.format(df_games['user_score'].isna().sum()))
```

## Исследовательский анализ данных

### Анализ выхода игр по годам

```python
# определение функции для bar chart
def bar_chart(dataframe, grouping_column, total_column):
             
    # названия графиков
    chart_name={'year_of_release':'Количество вышедших игр по годам (шт.)'
               }

    # группировка
    df_filtered = dataframe.groupby([grouping_column])[total_column].count()          
                
    # barchart для искомого показателя
    ax = df_filtered.plot(kind='bar',
                          y=total_column,
                          title=chart_name[grouping_column],
                          figsize=(15,5),
                          legend=False,
                          color = ['#0AA1DD'],
                          xlabel = ""
                         )
    # настройка подписей баров
    ax.bar_label(ax.containers[0], label_type='edge', color='black', fontsize=12, padding=10, rotation=90)
    
    # настройка осей
    ax.spines['top'].set_visible(False) #сброс границ графика
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    plt.yticks([]) #сброс y-оси
```

```python
# барчарт 
plt.ylim(0,1800)
bar_chart(df_games, 'year_of_release', 'name')
```

💡 До 1993 года игр выходило крайне мало — несколько десятков в год. Однако с 1994 г., когда количество выпущенных игр удвоилось по сравнению с прошлым годом, начался достаточно интенсивный рост с достижением локального пика в 2008 г. — с изданными 1427 играми. Далее выпуск пошел на спад и по состоянию на 2016 г. составил 502 игры.    
Ввиду малого количества вышедших игр в 1980-1993 г. можно заключить что интерес для исследования представляют данные об играх после 1994 г. 

### Анализ продаж игр по платформам

Проанализируем, как менялись продажи по платформам. Сначал определим платформы с наибольшими суммарными продажами.

```python
# определение функции для bar chart
def bar_chart_h(dataframe, grouping_columns, total_column):
             
    # названия графиков
    chart_name={'platform':'Суммарные продажи игр по платформам (в млн. копий)'
               }

    # группировка
    df_filtered = dataframe.groupby(grouping_columns)[total_column].sum().sort_values()          
    
    # barchart для искомого показателя
    ax = df_filtered.plot(kind='barh',
                          y=total_column,
                          title=chart_name[grouping_columns],
                          figsize=(15,8),
                          width=0.8,
                          legend=False,
                          color = ['#0AA1DD'],
                          xlabel = ""
                         )
   
    # настройка подписей баров
    ax.bar_label(ax.containers[0], label_type='edge', color='black', fontsize=10, padding=10)
    
    # настройка осей
    ax.spines['top'].set_visible(False) #сброс границ графика
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    plt.xticks([]) #сброс х-оси
    
    # таблица-рейтинг платформ
    top_sales = df_filtered.sort_values(ascending=False).reset_index()[grouping_columns].tolist()
    
    return top_sales
```

```python
# барчарт 
plt.xlim(0,1500)
top_sales = bar_chart_h(df_games, 'platform', 'total_sales')
```

```python
# определение функции
def scatter_plot(dataframe, y, x):
    
    """
    dataframe - набор данных
    y - столбец с зависимыми переменными (y-ось)
    x - столбец с независимыми переменная (x-ось)
    """
    
    # корреляция Пирсона
    display(dataframe[[y, x]].corr())

    # график зависимости
    dataframe.plot(kind='scatter',
                     y=y,
                     x=x,
                     title=f'Зависимость {y} от {x}', 
                     figsize=(15,5),
                     fontsize=12,
                     legend=True)
```

```python
# определение функции для bar chart
def bar_chart_combined(dataframe, grouping_column):
    
    ax = plt.gca()
    
    # группировка
    df_filtered = dataframe.groupby([grouping_column]).agg(
        {'name': 'count', 'total_sales': 'sum'}).round(0).sort_values(
        by='name', ascending=False)
       
    # barcharts
    barchart_1 = df_filtered['name'].plot(kind='bar',
                     figsize=(15,5),
                     color = '#0AA1DD',
                     width=0.3,
                     ax=ax,
                     position=1
                    )
    
    barchart_2 = df_filtered['total_sales'].plot(kind='bar',
                     figsize=(15,5),
                     color = '#F76E11',
                     width=0.3,
                     ax=ax,
                     position=0
                    )
    
    # настройка подписей баров
    ax.bar_label(ax.containers[0], label_type='edge', color='black', fontsize=10, padding=10, rotation=0)
    ax.bar_label(ax.containers[1], label_type='edge', color='black', fontsize=10, padding=10, rotation=0)
    
    # настройка осей
    ax.spines['top'].set_visible(False) #сброс границ графика
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    ax.legend(['Количество вышедших игр (шт.)', 'Суммарные продажи игр (млн. копий)']) #легенда
    ax.set_title('Количество вышедших игр (шт.) и суммарные продажи игр по жанрам (млн. копий)', fontsize=12) #настройка заголовка
    plt.yticks([]) #сброс y-оси
```

## Портрет пользователя в каждом регионе

Определим для пользователей регионов (NA, EU, JP) cамые популярные платформы и жанры (топ-5), опишем различия в долях продаж, исследуем влияние рейтинга ESRB на продажи в регионах.

```python
# определение функции для pie chart
def pie_chart(dataframe, grouping_columns, region, ax):
    
    # названия графиков
    chart_name={'na_sales':'Продажи в Сев. Америке',
                'eu_sales':'Продажи в Европе',
                'jp_sales':'Продажи в Японии'}
    
    chart_name_add={'platform': ' по топ-5 платформам',
                    'genre':' по топ-5 жанрам',
                    'rating':' по рейтингам'}
        
    # группировка
    df_filtered = dataframe.groupby([grouping_columns]).agg({region:'sum'}).sort_values(by=region, ascending=False).head()
    
    # pie_chart
    chart = df_filtered.plot(kind='pie',
                     ax=axes[ax],
                     y=region,
                     autopct=lambda p: '{:.2f}%\n({:.0f})'.format(p,(p/100)*df_filtered[region].sum()),
                     title=chart_name[region] + chart_name_add[grouping_columns],
                     fontsize=12,
                     legend=False,
                     colors = sns.color_palette("Blues_r"), #_r - reversed color
                     ylabel = "")
```

```python
# кортежу из двух элементов figure и axes передаем кол-тво строк/столбцов сетки подграфика
fig, axes = plt.subplots(1, 3, figsize=(18,5))

# pie_charts
pie_chart(df_games, 'platform', 'na_sales', 0)
pie_chart(df_games, 'platform', 'eu_sales', 1)
pie_chart(df_games, 'platform', 'jp_sales', 2)
```

💡 В Северной Америке в 2004-2016 гг. были наиболее популярными игры приставок Wii (22.6% продаж) и X360 (27.6%). В Европе PS3 с долей 27.5% продаж оказалась чуть популярнее располагающихся рядом друг с другом Wii и X360 с долями 21.7% и 22.6% соответственно.

## Проверка гипотез

Для проверки гипотез о равенстве средних двух генеральных совокупностей по взятым выборкам мы будем использовать t-тест.
Пороговое значение уровня значимости примем 0.05.

### Гипотеза о равенстве средних пользовательских рейтингов платформ Xbox One и PC

**Нулевая гипотеза (H0)** — средние пользовательские рейтинги платформ Xbox One и PC одинаковы.    

**Альтернативная гипотеза (H1)** — существуют значимые различия в средних пользовательских рейтингов платформ Xbox One и PC.

Проверим гипотезы о равенстве средних двух генеральных совокупностей по взятым из них выборкам:

```python
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергаем нулевую гипотезу
alpha = .05
```

```python
# формирование датафреймов
df_games_XOne = df_games.query('(platform == "XOne") & (user_score > 0)')
df_games_PC = df_games.query('(platform == "PC") & (user_score > 0)')
```

```python
# проверка гипотезы
results_1st_hyp = st.ttest_ind(df_games_XOne['user_score'], df_games_PC['user_score'])

print('p-value:', results_1st_hyp.pvalue)

if results_1st_hyp.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

💡 P-value меньше критического уровня статистической значимости — вероятность случайно получить различия в средних генеральных совокупностей составляет 0.1%, следовательно мы отвергаем нулевую гипотезу и можем сделать вывод о наличии статистически значимых различий средних пользовательских рейтингов платформ Xbox One и PC.

### Гипотеза о равенстве средних пользовательских рейтингов жанров Action и Sports

**Нулевая гипотеза (H0)** — средние пользовательские рейтинги жанров Action и Sports одинаковы.    

**Альтернативная гипотеза (H1)** — существуют значимые различия в средних пользовательских рейтингов жанров Action и Sports.

Проверим гипотезы о равенстве средних двух генеральных совокупностей по взятым из них выборкам:

```python
# формирование датафреймов
df_games_action = df_games.query('(genre == "Action") & (user_score > 0)')
df_games_sports = df_games.query('(genre == "Sports") & (user_score > 0)')
```

```python
# проверка гипотезы
results_2st_hyp = st.ttest_ind(df_games_action['user_score'], df_games_sports['user_score'])

print('p-value:', results_2st_hyp.pvalue)

if results_2st_hyp.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

💡 P-value меньше критического уровня статистической значимости, следовательно мы отвергаем нулевую гипотезу и можем сделать вывод о наличии статистически значимых различий средних пользовательских рейтингов жанров Action и Sports.
