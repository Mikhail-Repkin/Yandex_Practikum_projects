# Исследование надежности заемщиков.

**Описание проекта и цель исследования**

Заказчик — кредитный отдел банка. Входные данные от банка — статистика о платёжеспособности клиентов.

Цель исследования — проверьте гипотезу: 

* Влияет ли семейное положение и количество детей клиента на факт погашения кредита в срок.

Результаты исследования будут учтены при построении модели кредитного скоринга — специальной системы, которая оценивает способность потенциального заёмщика вернуть кредит банку.



**Ход исследования**

Данные о поведении пользователей получаем из файла `data.csv`. Перед проверкой гипотезы делаем обзор данных с целью проверить качество данных. 

Проверяем данные на ошибки и аномалии и оцениваем их влияние на исследование. Затем, на этапе предобработки выбираем способ обработки аномальных значений и исправляем самые критичные ошибки данных. Обрабатываем пропуски и удаляем дубликаты. Корректируем тип данных в столбцах.

Проводим категоризацию данных по доходам и целям кредита для последующей проверки гипотезы.
 
Таким образом, исследование пройдёт в три этапа:
 1. Обзор данных.
 2. Предобработка данных.
 3. Категоризация данных и проверка гипотезы.

### Шаг 2.1 Заполнение пропусков

```python
# подсчёт пропусков
df.isna().sum()
```

```python
# вывод на экран строк с пропущенными значениями в столбце total_income
df[df['total_income'].isna()].head()
```

```python
# подсчет доли пропущенных значений
missing_values_days = df['days_employed'].isna().sum()/len(df['days_employed'])
missing_values_income = df['total_income'].isna().sum()/len(df['total_income'])
print(f'Доля пропущеннных значений в столбце days_employed составляет: {missing_values_days:.1%}')
print(f'Доля пропущеннных значений в столбце total_income составляет: {missing_values_income:.1%}')
```

```python
# описательная статистика данных
print('Максимум:','{:.0f}'.format(df['total_income'].max()))
print('Минимум:','{:.0f}'.format(df['total_income'].min()))
print('Среднее арифметическое:','{:.0f}'.format(df['total_income'].mean()))
print('Медиана:','{:.0f}'.format(df['total_income'].median()))
```

```python
# заполнение пропусков в столбце total_income медианным значением (в зависимости от типа занятости)
df['total_income'] = df['total_income'].fillna(df.groupby('income_type')['total_income'].transform("median"))
df.isna().sum()
```

Все пропуски в столбце `total_income` заполнены медианным значением.

**Выводы**
* Кроме пропусков в столбце `days_employed` были обнаружены пропуски в столбце `total_income`. Все пропуски находятся в столбцах с количественными переменными и имеют тип `NaN`.
* Столбцы содержат одинаковое количество пропущенных значений. Доля пропусков в обоих столбцах составляет *10.1%*.
* Возможные причины появления пропусков в данных:   
   1) Пропуски в `total_income` не являются случайными — количество пропусков совпадает с количество пропусков в `days_employed`, следовательно пропуски в этих столбцах связаны. Возможной причиной может быть то, что потенциальный заёмщик не указал место работы и доход в анкете, либо сделал это неправильно.  
   2) Еще одной  причиной может быть сбой выгрузки данных из ПФР, либо внутренняя ошибка базы данных.  
   3) Также пропуски могут быть спецификой ведения базы данных — пока банк не получил документальное подтверждение дохода, выгрузка данных по трудовому стажу не происходит.
* В `total_income` содержится информация о доходах в виде количественных переменных (float64). Для данного типа данных лучший способ заполнения пропусков — медианные значения. Применение среднего значения привело бы к искажениям, а именно к завышению показателя, характеризующего выборку, т.к. данные по доходам имеют выбросы — максимальный доход составляет 2265604, минимальный — 20667,  когда как медиана — 145018.

```python
# вывод уникальных значений в столбцах
for column in df.columns:
    print()
    print(f'{column}:',df[column].unique())
```

В следующих столбцах были обнаружены артефакты (аномалии):  
1) `children`:    
- отрицательное количество детей (-1);  
- неправдоподобно большое количество детей (20).  

2) `days_employed`:  
- отрицательное количество дней трудового стажа;
- огромные значения положительных значений трудового стажа — максимальное значение 1100 лет;
- вещественный тип данных;
- пропуски.
   
3) `dob_years`:    
- нулевые значения возраста клиента.  
   
4) `education`:    
- дубликаты в категориальных переменных.  

5) `family_status`:  
- различный регистр.  
   
6) `gender`:

8) `purpose`:    
- дубликаты. 

**Обработка артефактов столбца `children`**

```python
# количество артефактов
df['children'].value_counts()
```

Всего в столбце 123 аномальных значения. Возможная причина артефактов — ошибка при вводе. Заменим значение -1 на 1, а значение 20 на 2.

```python
df.loc[df['children'] == -1,'children'] = 1
df.loc[df['children'] == 20,'children'] = 2
print(df['children'].unique())
```

В столбце не осталось аномальных значений

**Обработка артефактов столбца `days_employed`**

Возможная причина огромных значений трудового стажа — нарушилась разрядность в значениях. Разделим положительные значения на 100 и получим значения одного порядка с отрицательными:

```python
# нормирование положительных значений
df.loc[df['days_employed'] >0,'days_employed'] /= 100
```

```python
# проверка порядка значений
print(df.loc[df['days_employed'] >0,'days_employed'].mean())
print(df.loc[df['days_employed'] <0,'days_employed'].mean())
```

Теперь значения в столбце одного порядка.

Возможная причина отрицательных значений — перепутаны столбцы начала и окончания трудового стажа. Заменим отрицательные значения на положительные:

```python
# подсчет количества отрицательных значений
df[df['days_employed'] <0]['days_employed'].count()
```

```python
# замена отрицательных значений на положительные
df.loc[df['days_employed'] <0,'days_employed'] = df.loc[df['days_employed'] <0,'days_employed'].abs()
```

```python
# проверка замены значений
df[df['days_employed'] <0]['days_employed'].count()
```

Отрицательные значения заменены успешно.

```python
# заполнение пропусков в столбце медианным значением
df['days_employed'] = df['days_employed'].fillna(df['days_employed'].median())
df.isna().sum()
```

Все пропуски в столбце заполнены медианным значением.

```python
# замена вещественного типа данных в столбце на целочисленный
df['days_employed'] = df['days_employed'].astype('int64')
```

```python
df.head()
```

```python
df.info()
```

Теперь значения в `days_employed` корректы — отрицательные и аномально огромные значения отсутствуют, а пропуски заменены медианным значением. Вещественный тип данных в столбце заменен на целочисленный, а количество значений совпадает с первоначальным — все операции проведены успешно.

**Обработка артефактов столбца `dob_years`**

```python
# количество артефактов
df[df['dob_years'] == 0]['dob_years'].count()
```

Возможная причина нулевых значений возраста клиента — ошибка при вводе или загрузке данных. Заменим нулевые значения на медианные для групп клиентов по типу занятости:

```python
# расчет медианного возраста для клиентов сгруппированных по типу занятости
dob_years_med = df[df['dob_years'] != 0].groupby('income_type')['dob_years'].median().astype('int')
dob_years_med
```

```python
# замена нулевых значений на медианные
for types,med in dob_years_med.items():
    df.loc[(df['dob_years'] == 0) & (df['income_type'] == types), 'dob_years'] = med
```

```python
# проверка количества артефактов
df[df['dob_years'] == 0]['dob_years'].count()
```

Все нулевые значения в столбце `dob_years` заполнены соответствующим медианным значением.

**Обработка артефактов столбца `family_status`**

Возможная причина появления переменной с отличающимся регистром — ошибка ручного ввода. Приведем категориальные переменные к единому регистру:

```python
# приведение к нижнему регистру
df['family_status'] = df['family_status'].str.lower()
```

```python
# проверка регистра
df.sort_values(by='family_status')['family_status'].unique()
```

**Обработка артефактов столбца `gender`**

```python
# количество артефактов
df['gender'].value_counts()
```

Строка с пропуском только одна — удалим её.

```python
# удаление пропуска из датафрейма
df = df[df['gender'] != "XNA"]
print(df['gender'].value_counts())
df.shape
```

Строка удалена. Оставшее количество строк в датафрейме: 21524

**Выводы**  

На этапе предобработки данных были проведены следующие операции:
- Были обработаны основные артефакты в столбцах `children`, `days_employed`;
- Все пропуски в столбце `days_employed` заполнены медианным значением;
- Нулевые данные в столбце `dob_years` заполненны медианным значением согласно группировке клиентов по типу занятости;
- Категориальные переменные в `family_status` приведены к нижнему регистру;
- Удалена единственная строка с пропуском в столбце `gender`

Так как столбец `purpose` заполняется в свободной форме, на данном этапе не будем приводить его к унифицированному виду, вместо этого далее классифицируем цели кредита в отдельном столбце.

### Шаг 2.4. Удаление дубликатов.

```python
# проверка на отсутствие дубликатов
df.duplicated().sum()
```

```python
# приведение к нижнему регистру
df['education'] = df['education'].str.lower()
```

```python
# проверка регистра
df.sort_values(by='education')['education'].unique()
```

**Выводы**

- Возможными причинами появления дубликатов могут быть как технические ошибки так и человеческий фактор.

### Шаг 2.5. Формирование дополнительных датафреймов словарей, декомпозиция исходного датафрейма.

Создадим новый датафрейм, в котором каждому уникальному значению из `education` соответствует уникальное значение `education_id` 

```python
# создание нового датафрейма
education_dict = df[['education','education_id']]
```

```python
# удаление дубликатов из нового датафрейма
education_dict = education_dict.drop_duplicates().reset_index(drop=True)
education_dict
```

Создадим новый датафрейм, в котором каждому уникальному значению из `family_status` соответствует уникальное значение `family_status_id`

```python
# создание нового датафрейма
family_status_dict = df[['family_status','family_status_id']]
```

```python
# удаление дубликатов из нового датафрейма
family_status_dict = family_status_dict.drop_duplicates().reset_index(drop=True)
family_status_dict
```

Удалим из исходного датафрейма столбцы `education` и `family_status`.

```python
# удаление столбцов
df = df[['children', 'days_employed', 'dob_years', 'education_id',
       'family_status_id', 'gender', 'income_type', 'debt',
       'total_income', 'purpose']]
```

Удалим из исходного датафрейма явные дубликаты, которые образовались после удаления столбцов.

```python
# подсчет явных дубликатов
df.duplicated().sum()
```

### Шаг 2.6. Категоризация дохода.

Создадим столбец `total_income_category` с категориями:

0–30000 — `E`;\
30001–50000 — `D`;\
50001–200000 — `C`;\
200001–1000000 — `B`;\
1000001 и выше — `A`.  


Для этого напишем функцию категоризации клиентов по уровню дохода

```python
# функция категоризации клиентов по уровню дохода
def income_sorting(income):
    if income <= 30000:
        return 'E'
    if 30001 <= income <= 50000:
        return 'D'
    if 50001 <= income <= 200000:
        return 'C'
    if 200001 <= income <= 1000000:
        return 'B'
    else:
        return 'A'
```

```python
# применение функции к столбцу 'total_income'
df['total_income_category'] = df['total_income'].apply(income_sorting)
```

```python
# проверка работы функции
df.head()
```

### Шаг 2.7. Категоризация целей кредита.

Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории: 

 `операции с автомобилем`;\
 `операции с недвижимостью`;\
 `проведение свадьбы`;\
 `получение образования`.  

```python
# изучение данных в столбце 'purpose'
df.sort_values(by='purpose')['purpose'].unique()
```

```python
df.head()
```

```python
# функция категоризации цели получения кредита
def purpose_sorting(purpose):
    if  'авто' in purpose:
        return 'операции с автомобилем'
    elif 'недвиж' in purpose or 'жиль' in purpose:
        return 'операции с недвижимостью'
    elif 'свадьб' in purpose:
        return 'проведение свадьбы'
    elif 'образо' in purpose:
        return 'получение образования'
    else:
        return 'unknown'
```

```python
# применение функции к столбцу 'purpose'
df['purpose_category'] = df['purpose'].apply(purpose_sorting)
```

```python
# проверка работы функции
df.head()
```
